```python
# Let's investigate which month do the fires are more frequent?
#Greece
df_greece['acq_date'] = pd.to_datetime(df_greece['acq_date'])  # Ensure dates are in datetime format
df_greece['month'] = df_greece['acq_date'].dt.month  # Extract month from the date

# Group by month and count occurrences
monthly_fires = df_greece.groupby('month').size()

# Plotting
plt.figure(figsize=(10, 6))
monthly_fires.plot(kind='bar', color='orange')
plt.title('Monthly Fire Occurrences')
plt.xlabel('Month')
plt.ylabel('Number of Fires')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```
